<p># link structure</p>
"https://football-data.co.uk/mmz4281/" + "2526" + "/" + "E0" + ".csv"

```python
root = "https://football-data.co.uk/mmz4281/"
```

```python
leagues = ["E1","E2","E3","E0"]
frames = []
# appending the dataframe for leagues that got through pattern in frames[]
for category in leagues:
    df = pd.read_csv(root + '2526' + '/' + category + ".csv",encoding_errors='ignore')
    frames.append(df)
```

```python
# checking how many links
size = len(frames)
# checking first csv
frames[0].head()
```

<h2>Getting seasons</h2>

```python
leagues = ["E1","E2","E3","E0"]
frames = []
# appending the season dataframe that got through pattern in frames[]
for category in leagues:
    for seasons in range(20,24):
        df = pd.read_csv(root + str(seasons)+str(seasons+1) + '/' + category + ".csv",encoding_errors='ignore')
        # inserting column for identify season
        df.insert(1,'seasons',seasons)
    frames.append(df)
```

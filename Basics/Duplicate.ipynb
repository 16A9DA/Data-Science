```python
import pandas as pd
```

```python
df = pd.read_csv("laptop_price.csv",encoding_errors="ignore")
df.head()
```

```python
# duplicate in col/series
df[df.duplicated('laptop_ID')]
# @return no data duplicated
```

```python
df[df.duplicated(['Product','TypeName','Inches'])]
```

```python
# sorting data
df = df.sort_values(['Company','Price_euros'])
```

```python
df.value_counts('Company')
```

```python
duplicated_first = df.duplicated('Company')
```

```python
df[duplicated_first]
# cheapest laptop of each company ~ means not
df[~duplicated_first][['Company','Price_euros']]
# checking if any dup value
df[~duplicated_first][['Company','Price_euros']].value_counts('Company')
```

```python
duplicated_last = df.duplicated('Company',keep='last')
df[~duplicated_last][['Company','Price_euros']]
```

```python
duplicated_false = df.duplicated(['Company'],keep=False)
```

```python
# to get non duplicated false
df[~duplicated_false]
```

```python
# dropping duplicates
df.drop_duplicates(['Company'])
df.drop_duplicates(['Company']).value_counts(['Company'])
```

```python
df = df.sort_values(by=['Company','Price_euros'])
```

```python
# cheapest laptop
df.drop_duplicates(['Company'],keep='first')
```

```python
df = pd.read_csv('laptop_price.csv',encoding_errors='ignore')
#finding big and small size in laptop using sort and duplicate first,last and false
df = df.sort_values(['Company','Inches'])
df.drop_duplicates(['Company'],keep='first')[['Company','Price_euros']]

df.drop_duplicates(['Company'],keep='last')[['Company','Price_euros']]
```

```python
import pandas as pd, matplotlib.pyplot as plt
```

```python
# raw population data/unedited
df = pd.read_csv("population_total.csv",encoding_errors="ignore")
df
```

```python
df.dropna(inplace=True)
```

```python
df = df.pivot(index="year",columns="country",values="population")
```

```python
df_pivot = df[["United States","India","China","Indonesia","Brazil"]]
```

```python
# lineplot
df_pivot.plot(kind="line",xlabel="Year",ylabel="Population")
plt.savefig("Test.png")
plt.show()
```

```python
# barplot, one row only,country names in index for barplot
df_pivot20 = df_pivot[df_pivot.index.isin([2020])]
df_pivot20.T.plot(kind="bar",color="orange",xlabel="Country",ylabel="Population",title="Population 2020")
```

```python
df_pivot
df_bar = df_pivot[df_pivot.index.isin([1980,1990,2000,2010])]
df_bar.plot(kind="bar")
```

```python
# pie chat
df_pivot20 = df_pivot20.T
df_pivot20 = df_pivot20.rename(columns={2020.0:'2020'})
df_pivot20.plot(kind="pie",y="2020")
```

```python
# boxplot
df_pivot.plot(kind="box")
```

```python
df_pivot.plot(kind="hist")
```

```python
df = pd.read_csv("population_total.csv",encoding_errors="ignore")
df_sample = df[df["country"].isin(["United States","Brazil"])]
df_sample.plot(kind="scatter",x="year",y='population',s=80,color="green")
```

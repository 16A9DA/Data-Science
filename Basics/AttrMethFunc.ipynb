```python
# adding column
# adding elements different to dataframe
import numpy as np
df.info()
# 1000 elements 
language_score = np.arange(0,1000)
len(language_score)
df["Language Score"] = language_score 
df.head(5)
```

```python
# random int grades from 0 to 100
# size for 1000 elements
# 0 is inclusive in random, 100 exclusive similar to range
scores =np.random.randint(0,100,size=1000)
len(scores)
df["Language Score"] = scores
df.head()
min(df["Language Score"])
```

```python
# random float number 
scores = np.random.uniform(0,100,size=1000)
df["Language Score"] = scores
df.head()
```

```python
# for multiple column using assign
# creating new scores for new columns
score1 = np.random.randint(1,1000,size=1000)
score2 = np.random.randint(1,100,size=1000)
# creating columns - > series are rows
series1 = pd.Series(score1,index=np.arange(0,1000))
series2  = pd.Series(score2,index=np.arange(0,1000))
# Name of column and rows
# df = always to overwrite
df = df.assign(Score1=series1,Score2=series2)
df.head()

```

```python
# Using index to add a column at specific location we use insert
# insert(position,name,values)
# no need for df = in insert
df.insert(1,"Test",series1)
```

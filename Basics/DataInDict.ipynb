```python
import pandas as pd
```

```python
root = "https://football-data.co.uk/mmz4281/"

leagues = ["E1","E2","E3","E0"]

frames = []
for category in leagues:
    for seasons in range(20,24):
        df = pd.read_csv(root + str(seasons)+str(seasons+1) + '/' + category + ".csv",encoding_errors='ignore')
        # inserting column for identify season
        df.insert(1,'seasons',seasons)
    frames.append(df)
```

```python
dict = {
    "English Premier League":"E0",
    "English League":"E1",
    "English League 1":"E2",
    "English League 2":"E3"  
}
```

```python
root = "https://football-data.co.uk/mmz4281/"
# we will get value from dict line55
# leagues = ["E1","E2","E3","E0"]
dict_historical_event = {}
for leagues in dict:
    for seasons in range(20,24):
        # we dont want repitition of data so, append and remove so we will put inside look
        frames = []
        df = pd.read_csv(root + str(seasons)+str(seasons+1) + '/' + dict[leagues] + ".csv",encoding_errors='ignore')
        # inserting column for identify season
        df.insert(1,'seasons',seasons)
    frames.append(df)
    # making all dataframe to 1 dataframe
    df_concat = pd.concat(frames)
    # making it a dict with val 
    dict_historical_event[leagues] = df_concat
```

```python
import re 
```

```python
text = """Hello World.
I love Python!!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456"""
```

```python
re.search(r'\d',text)
re.search(r'^hello',text,flags=re.IGNORECASE)
```

```python
re.findall(r"\d",text)
```

```python
text = """Hello World.
I love Python!!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456"""

text2 = """
13-04-2021
2021-13-04
2021-04-13
"""

text3 = """
username10
abc
10
"""
```

```python
# match only puntuation
re.findall(r'[^\w\s]',text)
```

```python
# match right format (MM-DD/-YYYY)
re.findall(r"\d{2}-\d{2}-\d{4}",text2)
```

```python
# match username format with 4-14 char and only letters and digit

re.findall(r"[a-zA-Z0-9]{4,14}",text3)
```
